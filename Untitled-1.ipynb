```python
#@jit(nopython=True)
def cosine(x,y):
    #normalized vectors
    x /= np.linalg.norm(x)
    y /= np.linalg.norm(y)
    return np.dot(x,y)

def cosine_vet(X,Y):
    X = X/np.linalg.norm(X, axis=1, keepdims=True)
    Y = Y/np.linalg.norm(Y, axis=1, keepdims=True)
    
    dot_product = np.dot(X,Y.T)
    
    return dot_product

def week(i):
        switcher={
                0:'Sunday',
                1:'Monday',
                2:'Tuesday',
                3:'Wednesday',
                4:'Thursday',
                5:'Friday',
                6:'Saturday'
             }
        return switcher.get(i,"Invalid day of week")
```

```python
print(f'Carregando imagens na memória')


import tkinter as tk
from tkinter import filedialog, messagebox
from tkinter.messagebox import showinfo

## testes
root = tk.Tk()
root.title('Testes...')
root.resizable(False, False)
root.geometry('300x150')
# run the application
root.mainloop()


file_pathP = filedialog.askopenfilename(title='Selecione a face padrão...');
assert len(file_pathP) > 0, "Não foi selecionada nenhuma imagem padrão!" 

file_pathQ = filedialog.askopenfilename(title='Selecione a face questionada...');
assert len(file_pathP) > 0, "Não foi selecionada nenhuma imagem questionada!" 


str_pathP = file_pathP
str_pathQ = file_pathQ
imageP = cv2.imread(str_pathP)
imageQ = cv2.imread(str_pathQ)
```

```python

# ajuda a visualizar melhor as faces e as marcações, tomando as dimensões pela menor imagem?
if imageP.shape < imageQ.shape:
    height, width = imageP.shape[:2]
    imageQ = cv2.resize(imageQ, (width, height), interpolation = cv2.INTER_CUBIC)
else:
    height, width = imageQ.shape[:2]
    imageP = cv2.resize(imageP, (width, height), interpolation = cv2.INTER_CUBIC)
```

```python
print('Verificando e extraindo features das faces...')

ff = ForensicFace(model='sepaelv2', det_size=320, use_gpu=True, gpu=0, extended=True, magface=False)

resultado1 = ff.process_image_multiple_faces(imageP)
resultado2 = ff.process_image_multiple_faces(imageQ)

if len(resultado1) == 1:
    print(resultado1[0].keys())
else:
    # print("revise a imagem padrão")
    fig = plt.figure(f'Revise a imagem Padrão. Encontradas {len(resultado1)} faces.' )
    fig.suptitle(f'\nFeche a janela para continuar...\n')
    
    for ki in resultado1:
        bbox = ki["bbox"]
        imageP = cv2.rectangle(imageP, pt1=(bbox[0], bbox[1]), pt2=(bbox[2], bbox[3]), thickness=2, color=(255,0,255))
                
    plt.imshow(cv2.cvtColor(imageP, cv2.COLOR_BGR2RGB), aspect='auto')
    plt.show(block='true');  # display it
    exit()
    
if len(resultado2) == 1:
        print(resultado2[0].keys())
else:
    fig = plt.figure(f'Revise a imagem Questionada. Encontradas {len(resultado2)} faces.')
    fig.suptitle('\nFeche a janela para continuar...\n')
    
    for ki in resultado2:
        bbox = ki["bbox"]
        imageQ = cv2.rectangle(imageQ, pt1=(bbox[0], bbox[1]), pt2=(bbox[2], bbox[3]), thickness=2, color=(255,0,255))
        
    plt.imshow(cv2.cvtColor(imageQ, cv2.COLOR_BGR2RGB), aspect='auto')
    plt.show(block='true');  # display it
    exit()
    
```

```python
ids = df.identity.to_numpy()
X = np.array(df.adaface_features.to_list())

X.shape, ids.shape
```

```python
escores = cosine_vet(X, X)
same_id_mask = ids[:,np.newaxis] == ids
ss = escores[same_id_mask]
ss = ss[ss<0.99]
ds = escores[~same_id_mask]
print(f"Quantidade de escores SS: {len(ss)}\nQuantidade de escores DS: {len(ds)}")
```

```python
plt.rcParams["figure.figsize"] = (16,8)
plt.rcParams["font.family"] = 'sans-serif' #'FreeSerif'
plt.rcParams["font.size"] = 18


xmin=-0.3
xmax=1.
n=100
x = np.linspace(start=xmin, stop=xmax, num=n, endpoint=True)

ss_h, _ = np.histogram(ss, bins=n, range=(xmin,xmax), density=True)
ds_h, _ = np.histogram(ds, bins=n, range=(xmin,xmax), density=True)

max_hist = max(max(ss_h),max(ds_h))
y_lim = max_hist*1.1

plt.bar(x,ss_h, width=0.6*(xmax-xmin)/len(x), alpha=0.5, linewidth=.5, edgecolor='b', label="Mesma pessoa")
plt.bar(x,ds_h, width=0.6*(xmax-xmin)/len(x), alpha=0.5, linewidth=.5, edgecolor='orange', label="Pessoas diferentes")

#for score in scores:
#    plt.vlines(x=score, ymin=0, ymax=8, linestyle='dashed', linewidth=1, color='r', label=f'Escore = {score:.3f}')
# desenha linha vertical indicando o score da comparação atual
plt.vlines(x=simil, ymin=0, ymax=8, linestyle='dashed', linewidth=1, color='r', label=f'Escore = {simil:.3f}')

plt.ylim(0,y_lim)
plt.grid(linestyle='dotted', color='k', linewidth=0.2)
plt.xlabel('Escores')
plt.ylabel('Freq. relativa')
plt.legend()
```
